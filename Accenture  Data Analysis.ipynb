```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Content.csv')
df
```

```python
df.isnull().sum()
```

```python
df2 = pd.read_csv('Reactions.csv')
df2
```

```python
df2.isnull().sum()
```

```python
df3 = pd.read_csv('ReactionTypes.csv')
df3
```

```python
df3.isnull().sum()
```

```python
# merge the reaction table with the content table

merged_df = pd.merge(df2, df, left_on="ContentID", right_on="Content ID", how="left")
merged_df
```

```python
# merge the merged_df with reaction type table

final_df = pd.merge(merged_df, df3, left_on="Reaction type", right_on="Type", how="left")
final_df
```

```python
# droping unnecessary columns

final_df.drop(['Unnamed: 0_x', 'Content ID', 'Unnamed: 0_y', 'Type'], axis=1, inplace=True)
final_df
```

```python
#Figure out the Top 5 performing categories Add up the total scores for each category.

#Group the data by category and sum the score

category_scores = final_df.groupby('Category')['Score'].sum()
category_scores
```

```python
# sort the category based on total score in descending order

sorted_category = category_scores.sort_values(ascending=False)
sorted_category
```

```python
#Get the top 5 performing categories

top_5_category = sorted_category.head(5)
top_5_category
```

```python
new_df = final_df.copy()
```

```python
new_df.to_csv('cleaned_data_with_top_categories.csv', index=False)
```

```python
import matplotlib.pyplot as plt
```

```python
# Number of unique categories

num_unique_categories = len(final_df['Category'].unique())
num_unique_categories
```

```python
# create bar chart 

plt.bar('Unique Categories', num_unique_categories)
# Set labels and title
plt.xlabel('Insight')
plt.ylabel('Count')
plt.title('Number of Unique Categories')

# Display the chart
plt.show()
```

```python
# Count the number of reactions for each category
reaction_counts = final_df['Category'].value_counts()

# Get the most popular category
most_popular_category = reaction_counts.index[0]

# Get the number of reactions for the most popular category
num_reactions_most_popular = reaction_counts.iloc[0]

# Create a pie chart to visualize the number of reactions for the most popular category
plt.pie([num_reactions_most_popular, len(final_df) - num_reactions_most_popular], labels=['Most Popular', 'Others'], autopct='%1.1f%%')

# Set title
plt.title('Reactions for the Most Popular Category')

# Display the chart
plt.show()
```

```python
import pandas as pd

# Convert the 'Datetime' column to datetime type
final_df['Datetime'] = pd.to_datetime(final_df['Datetime'])

# Extract the month from the 'Datetime' column
final_df['Month'] = final_df['Datetime'].dt.month

# Count the number of posts for each month
month_counts = final_df['Month'].value_counts()

# Get the month with the most posts
most_posts_month = month_counts.idxmax()

# Create a bar chart to visualize the number of posts for each month
plt.bar(month_counts.index, month_counts.values)

# Set labels and title
plt.xlabel('Month')
plt.ylabel('Number of Posts')
plt.title('Number of Posts by Month')

# Display the chart
plt.show()
```
